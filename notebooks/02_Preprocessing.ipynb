```python
# Import pandas for data manipulation
import pandas as pd

# Import NumPy for numerical operations
import numpy as np
```

```python
# Read the CSV file
df = pd.read_csv(r'../dataset/mymoviedb.csv', engine='python', on_bad_lines='skip', encoding='utf-8')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Convert to numeric, coerce errors to NaN
df['Vote_Average'] = pd.to_numeric(df['Vote_Average'], errors='coerce')

# Also convert Vote_Count while we're at it
df['Vote_Count'] = pd.to_numeric(df['Vote_Count'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#df["Overview"] = df["Overview"].fillna("")
```

```python
# ==========================================
# Remove Duplicate Records
# ==========================================

print("Before :", df.shape)

df.drop_duplicates(inplace=True)

print("After :", df.shape)
```

```python
df["Release_Date"] = pd.to_datetime(
    df["Release_Date"],
    errors="coerce"
)
```

```python
df.info()
```

```python
df["Year"] = df["Release_Date"].dt.year
```

```python
df.head()
```

```python
df["Genre"] = df["Genre"].str.lower()

df["Overview"] = df["Overview"].str.lower()

df["Original_Language"] = df["Original_Language"].str.lower()
```

```python
df["Genre"] = df["Genre"].str.strip()

df["Overview"] = df["Overview"].str.strip()

df["Original_Language"] = df["Original_Language"].str.strip()
```

```python
df["Tags"] = (
    df["Genre"] + " " +
    df["Overview"] + " " +
    df["Original_Language"]
)
```

```python
df["Tags"] = df["Tags"].str.replace(r"\s+", " ", regex=True)
```

```python
df[["Title","Tags"]].head()
```

```python
# ==========================================
# Final Dataset for Recommendation System
# ==========================================

movies = df[
    [
        "Title",
        "Tags",
        "Genre",
        "Overview",
        "Original_Language",
        "Popularity",
        "Vote_Average",
        "Vote_Count",
        "Poster_Url",
        "Year"
    ]
].copy()
```

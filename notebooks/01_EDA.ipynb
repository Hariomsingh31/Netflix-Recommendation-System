```python
# Import required libraries

import pandas as pd
import numpy as np

# Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Display all columns
pd.set_option('display.max_columns', None)

# Set graph style
plt.style.use("ggplot")



```

```python
df = pd.read_csv(r'../dataset/mymoviedb.csv', engine='python', on_bad_lines='skip', encoding='utf-8')
```

```python
# Display first 5 rows
df.head()
```

```python
# Convert to numeric, coerce errors to NaN
df['Vote_Average'] = pd.to_numeric(df['Vote_Average'], errors='coerce')

# Also convert Vote_Count while we're at it
df['Vote_Count'] = pd.to_numeric(df['Vote_Count'], errors='coerce')
```

```python
# Number of rows and columns

df.shape
```

```python
# Show all column names

df.columns
```

```python
df.info()

```

```python
df.describe()

```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(df.isnull(), cbar=False)

plt.title("Missing Values")

plt.show()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df["Release_Date"] = pd.to_datetime(df["Release_Date"], errors='coerce', format='%Y-%m-%d')
```

```python
df.info()
```

```python
df["Year"] = df["Release_Date"].dt.year
```

```python
df.head()
```

```python
print("Total Movies :",len(df))
```

```python
df["Original_Language"].nunique()
```

```python
df["Original_Language"].value_counts()
```

```python
plt.figure(figsize=(10,5))

df["Original_Language"].value_counts().head(10).plot(kind="bar")

plt.title("Top Languages")

plt.xlabel("Language")

plt.ylabel("Movies")

plt.show()
```

```python
genre = df["Genre"].str.split(",",expand=True)

genre.head()
```

```python
genres = (
df["Genre"]
.str.split(",")
.explode()
.str.strip()
)
```

```python
genres.value_counts()
```

```python
plt.figure(figsize=(10,6))

genres.value_counts().head(10).plot(kind="bar")

plt.title("Top Genres")

plt.show()
```

```python
top_rating = df.sort_values(
"Vote_Average",
ascending=False
)

top_rating[
["Title","Vote_Average"]
].head(10)
```

```python
popular = df.sort_values(
"Popularity",
ascending=False
)

popular[
["Title","Popularity"]
].head(10)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
x="Popularity",
y="Title",
data=popular.head(10)
)

plt.title("Top 10 Popular Movies")

plt.show()
```

```python
vote = df.sort_values(
"Vote_Count",
ascending=False
)

vote[
["Title","Vote_Count"]
].head(10)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
df["Vote_Average"],
bins=20
)

plt.title("Vote Average Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
df["Popularity"],
bins=20
)

plt.title("Popularity Distribution")

plt.show()
```

```python
plt.figure(figsize=(14,6))

df["Year"].value_counts().sort_index().plot()

plt.title("Movies Released by Year")

plt.xlabel("Year")

plt.ylabel("Movies")

plt.show()
```

```python
top10 = df.sort_values(
"Popularity",
ascending=False
).head(10)

plt.figure(figsize=(12,6))

sns.barplot(
data=top10,
x="Popularity",
y="Title"
)

plt.title("Top 10 Popular Movies")

plt.show()
```

```python

numeric = df.select_dtypes(include=np.number)

plt.figure(figsize=(8,5))

sns.heatmap(
numeric.corr(),
annot=True,
cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```
